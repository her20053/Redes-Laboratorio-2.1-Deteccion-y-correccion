```python
import matplotlib.pyplot as plt
import numpy as np

with open("simulacion.txt", "r") as file:
    
    content = file.readlines()

# Convert the content into a list of integers
percentages = [int(line.strip()) for line in content]

# 1. Histogram
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.hist(percentages, bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de Porcentajes')
plt.xlabel('Porcentaje')
plt.ylabel('Frecuencia')

# 2. Boxplot
plt.subplot(1, 3, 2)
plt.boxplot(percentages, vert=False)
plt.title('Boxplot de Porcentajes')
plt.yticks([])

# 3. Cumulative Distribution Function (CDF)
sorted_percentages = np.sort(percentages)
cdf = np.arange(1, len(sorted_percentages)+1) / len(sorted_percentages)
plt.subplot(1, 3, 3)
plt.plot(sorted_percentages, cdf, marker='.', linestyle='none', color='green')
plt.title('Función de Distribución Acumulada')
plt.xlabel('Porcentaje')
plt.ylabel('Probabilidad')

plt.tight_layout()
plt.show()
```
